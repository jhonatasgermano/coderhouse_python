```python
import pandas as pd
```

```python
# Abra o arquivo "best-selling game consoles.xlsx" em um dataframe.
```

```python
df = pd.read_excel("best-selling game consoles.xlsx")
df.head()
```

```python
# Substitua a palavra "NES" por "Nitendinho" no nome dos consoles e deixe todos os nomes em mai√∫sculos.
```

```python
df['Console Name'] = df['Console Name'].str.replace('NES', 'Nintendinho').str.upper()
```

```python
# Filtre o nome dos consoles com data de release depois de 2010.
```

```python
df[df['Released Year'] > 2010]
```

```python
# De um describe e info da base, substitua os missing values pela string "missing".
```

```python
df['Remarks'] = df['Remarks'].fillna('missing')
df.head()
```

```python
df[(df['Discontinuation Year'] - df['Released Year'] < 2) & (df['Discontinuation Year'] - df['Released Year'] >= 0)]
```
